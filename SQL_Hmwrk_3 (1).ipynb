# **Sales and Revenue Analysis**

# Payment Types

```python
query = """SELECT payment_type, COUNT(payment_value) as Payments_Count FROM order_payments
GROUP BY payment_type
ORDER BY SUM(payment_value) DESC
;"""
pd.read_sql(query, conn)
```

```python
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(8, 5))
plt.bar(df['payment_type'], df['Payments_Count'], color='skyblue')
plt.title('Payment Type Distribution', fontsize=16)
plt.xlabel('Number of Payments')
plt.ylabel('Payment Type')
plt.show()
```

```python
query = """SELECT payment_type, AVG(payment_value) as AVG_payment FROM order_payments
GROUP BY payment_type
ORDER BY SUM(payment_value) DESC
;"""
pd.read_sql(query, conn)
```

# Revenue from Products sorted by Categories

```python
query = """SELECT  product_category_name_translation.product_category_name_english, SUM(price+freight_value) as Revenue FROM order_items
JOIN products ON order_items.product_id = products.product_id
JOIN product_category_name_translation ON products.product_category_name = product_category_name_translation.product_category_name
GROUP BY product_category_name_translation.product_category_name_english
;"""
pd.read_sql(query, conn)
```

# Top5-selling products

```python
query = """SELECT  product_category_name_translation.product_category_name_english as Mostly_sold_product_type, SUM(price+freight_value) as Revenue FROM order_items
JOIN products ON order_items.product_id = products.product_id
JOIN product_category_name_translation ON products.product_category_name = product_category_name_translation.product_category_name
GROUP BY product_category_name_translation.product_category_name_english
ORDER BY
    Revenue DESC
Limit 5
;"""
pd.read_sql(query, conn)
```

```python
df = pd.read_sql_query(query, conn)
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x='Mostly_sold_product_type', y='Revenue', palette='mako')

# Add titles and labels
plt.title('Top 5 Selling Product Types by Revenue', fontsize=14)
plt.xlabel('Product Type', fontsize=12)
plt.ylabel('Revenue (in $)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

plt.show()
```

Most of the sellers' locations

```python
query = """SELECT seller_city, COUNT(seller_zip_code_prefix) as NumberOfSellers FROM sellers
GROUP BY sellers.seller_city
ORDER BY
    NumberOfSellers DESC
Limit 10;"""
pd.read_sql(query, conn)
```

```python
df_Seller_cities = pd.read_sql_query(query, conn)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_Seller_cities, y='seller_city', x='NumberOfSellers', palette='viridis')

# Add titles and labels
plt.title('Top 10 Cities by Number of Sellers', fontsize=14)
plt.xlabel('Number of Sellers', fontsize=12)
plt.ylabel('City', fontsize=12)
```

# Customer data

```python
query = "SELECT * FROM customers LIMIT 5;"
pd.read_sql(query, conn)
```

```python
query = """SELECT customer_city, COUNT(customer_zip_code_prefix) as NumberOfCustomers FROM customers
GROUP BY customers.customer_city
ORDER BY
    NumberOfCustomers DESC
Limit 10;"""
pd.read_sql(query, conn)
```

```python
df_cust_cities = pd.read_sql_query(query, conn)
plt.figure(figsize=(10, 6))
sns.barplot(data=df_cust_cities, y='customer_city', x='NumberOfCustomers', palette='viridis')

# Add titles and labels
plt.title('Top 10 Cities by Number of Sellers', fontsize=14)
plt.xlabel('Number of Sellers', fontsize=12)
plt.ylabel('City', fontsize=12)
```

```python
query = """
SELECT
sellers.seller_city,
product_category_name_translation.product_category_name_english as product_type,
order_items.product_id
FROM sellers
JOIN order_items ON sellers.seller_id = order_items.seller_id
JOIN products ON order_items.product_id = products.product_id
JOIN product_category_name_translation ON products.product_category_name = product_category_name_translation.product_category_name
GROUP BY sellers.seller_city
;"""
pd.read_sql(query, conn)
```
