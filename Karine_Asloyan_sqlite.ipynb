1. Find all records where the total_bill exceeds $50. (2 points

```python
query = "SELECT * FROM Observation;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM Observation
WHERE Observation.total_bill > 50
;"""
pd.read_sql(query, conn)
```

2. Calculate the average tip given. (2 points)

```python
query = """
SELECT AVG(Observation.tip) as Average_tip FROM Observation
;"""
pd.read_sql(query, conn)
```

3. Find the maximum total_bill paid. (2 points)

```python
query = """
SELECT MAX(Observation.total_bill) as Maximum_bill FROM Observation
;"""
pd.read_sql(query, conn)
```

4. Count how many people were served on each day. (4 points)

```python
query = "SELECT * FROM Day;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Observation.size) as customers, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY Day.day
;"""
pd.read_sql(query, conn)
```

5. Calculate the average total_bill grouped by size

```python
query = """
SELECT AVG(Observation.total_bill) as Average_bill, Observation.size FROM Observation
GROUP BY  Observation.size
;"""
pd.read_sql(query, conn)
```

6. Find the total tip amount for smokers vs. non-smokers.

```python
query = "SELECT * FROM Smoker;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Observation.tip) as Total_tip, Smoker.smoker FROM Observation
JOIN Smoker ON Smoker.smoker_id=Observation.smoker_id
GROUP BY Smoker.smoker
;"""
pd.read_sql(query, conn)
```

7. Determine the average tip percentage (tip/total_bill) for each day. (4 points)

```python
query = """
SELECT (Observation.tip)*100/(Observation.total_bill) as Average_tip_perc, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY  Day.day
;"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT AVG((Observation.tip)*100/(Observation.total_bill)) as Average_tip_perc, Day.day FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
GROUP BY  Day.day
;"""
pd.read_sql(query, conn)
```

8. List all total_bill values with the corresponding day and time. (4 points)

```python
query = "SELECT * FROM Time;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT Observation.total_bill as Totall_bill, Day.day, Time.time FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
JOIN Time ON Time.time_id=Observation.time_id
GROUP BY Observation.total_bill
;"""
pd.read_sql(query, conn)
```

9. Find the total number of observations for each gender. (4 points)

```python
query = """
SELECT COUNT(Observation.sex_id) as Total_number, Sex.sex FROM Observation
JOIN Sex ON Sex.sex_id=Observation.sex_id
GROUP BY Sex.sex
;"""
pd.read_sql(query, conn)
```

10. Retrieve all tips given during Dinner on Saturday. (4 points)

```python
query = "SELECT * FROM Time;"
pd.read_sql(query, conn)
```

```python
query = """
SELECT SUM(Observation.tip) as Tip, Time.time FROM Observation
JOIN Time ON Time.time_id=Observation.time_id
JOIN Day ON Day.day_id=Observation.day_id
WHERE Day.day = 'Sat' AND Time.time = 'Dinner'
;"""
pd.read_sql(query, conn)
```


# **<font color='red'> All the records satisfying the defined condition should be displayed. Minus 1 point </font>**

11. Find the largest total_bill for each gender. (4 points)

```python
query = """
SELECT MAX(Observation.total_bill) as Largest_bill, Sex.sex FROM Observation
JOIN Sex ON Sex.sex_id=Observation.sex_id
GROUP BY Sex.sex
;"""
pd.read_sql(query, conn)
```

12.Find the average tip for parties of size greater than 4. (4 points)

```python
query = """
SELECT AVG(Observation.tip) as Av_tip FROM Observation
WHERE Observation.size > 4
;"""
pd.read_sql(query, conn)
```

13.Find all records from where the day is either Friday or Saturday, and the time is Lunch or Dinner. (4 points)

```python
query = """
SELECT * FROM Observation
JOIN Day ON Day.day_id=Observation.day_id
JOIN Time ON Time.time_id=Observation.time_id
WHERE (Day.day = 'Sat' OR Day.day = 'Fri') AND (Time.time = 'Lunch' OR Time.time = 'Dinner')
;"""
pd.read_sql(query, conn)
```

14.List all records where the tip percentage is higher than 20%. (4 points)

```python
query = """
SELECT * FROM Observation
WHERE (Observation.tip*100)/Observation.total_bill > 20
;"""
pd.read_sql(query, conn)
```

15. Use a subquery (query inside query) to find the record with the second highest tip. (you need to use a single query

```python
query = """
SELECT * FROM Observation
WHERE Observation.tip = (
    SELECT MAX(Observation.tip) FROM Observation
    WHERE Observation.tip < (
        SELECT MAX(Observation.tip) FROM Observation
    )
)
;"""
pd.read_sql(query, conn)
```
